```python
# %matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[1, 2],
             [2, 9]])

xi = np.arange(-2, 2, 0.1)

N = len(xi)

quad_form = np.zeros((N, N))
```

```python
for i in range(N):
    for j in range(N):

        x = np.array([xi[i], xi[j]]).T

        normfact = x.T@x

        quad_form[i, j] = x.T@A@x / normfact
```

```python
eig_val, eig_vec = np.linalg.eig(A)
```

```python
X, Y = np.meshgrid(xi, xi)

fig = plt.figure(figsize=(7, 7))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, quad_form, cmap='jet')

qf_max = quad_form.max() + 0.5
ax.plot([0, eig_vec[0, 0]], [0, eig_vec[0, 1]], np.array([qf_max, qf_max]), linewidth=5)
ax.plot([0, eig_vec[1, 0]], [0, eig_vec[1, 1]], [qf_max, qf_max], linewidth=5)

ax.set_xlabel('w1')
ax.set_ylabel('w2')
ax.set_zlabel('energy')
plt.show()
```
