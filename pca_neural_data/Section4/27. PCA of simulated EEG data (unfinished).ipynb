```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io

# %matplotlib notebook
```

```python
# index of diploe to simulate activity in
diploc = 109
```

```python
data = scipy.io.loadmat('emptyEEG.mat')
```

```python
grid_loc = data['lf']['GridLoc'][0, 0]
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(grid_loc[:, 0], grid_loc[:, 1], grid_loc[:, 2], 'bo', markerfacecolor='y')
ax.plot3D(grid_loc[diploc, 0], grid_loc[diploc, 1], grid_loc[diploc, 2], 'rs', markerfacecolor='k', markersize=10)
plt.title('Brain dipole locations')

plt.show()
```
